### Regression Analysis

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

from sklearn import preprocessing

import statsmodels.api as sm
```

```python
import json
```

```python
evaluate_cate = ['anger',
 'anticipation',
 'disgust',
 'fear',
 'joy',
 'negative',
 'positive',
 'sadness',
 'surprise',
 'trust']
evaluation = evaluate_cate[5]

agency = 'CFPB'
```

```python
f = open("./emotion_dict_" + agency + ".json","r")
```

```python
dict_OCC = json.load(f)
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
def get_emotion_df(evaluation, dict_OCC):
    dict_ = {}
    for key in dict_OCC.keys():
        empty_1 = 0
        try:
            rule_count = dict_OCC[key]['Rule'][evaluation][0]
        except:
            rule_count = 0
        pro_rule_count = 0
        comment_count = 0
        for pro_rule in dict_OCC[key]['Proposed Rule'].keys():
            try:
                pro_rule_count += dict_OCC[key]['Proposed Rule'][pro_rule][evaluation][0]
            except:
                pro_rule_count += 0
                empty_1 += 1

            try:
                comment_array = np.array(dict_OCC[key]['Proposed Rule'][pro_rule]['comment'][evaluation])
            except:
                comment_array = np.zeros(1)
            comment_array = comment_array[comment_array != 0]
            try:
                min_max_scaler.fit_transform(comment_array[:,np.newaxis])
                comment_count += np.std(comment_array)
            except:
                
                comment_count += np.std(comment_array)

        comment_count = comment_count / len(dict_OCC[key]['Proposed Rule'])
        try:
            pro_rule_count = pro_rule_count / (len(dict_OCC[key]['Proposed Rule']) - empty_1)
        except:
            pro_rule_count = 0

        dict_[key] = [rule_count, pro_rule_count, comment_count]
    
    df = pd.DataFrame.from_dict(dict_, orient='index', columns = ['Rule ' + evaluation, 
                                                                  'Proposed Rule ' + evaluation, 
                                                                  'Comment Std ' + evaluation])
        
    return df
```

```python
df = get_emotion_df(evaluation, dict_OCC)
```

```python
df = df.sort_values(by = 'Comment Std negative', ascending = False)
df
```

```python
x_factor = 'Proposed Rule ' + evaluation 
y_factor = 'Comment ' + evaluation 

scale = True
```

```python
df.replace(0, np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
x = np.array(df[x_factor])[:,np.newaxis]
y = np.array(df[y_factor])[:,np.newaxis]
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
if scale:
    x = min_max_scaler.fit_transform(x)
    y = min_max_scaler.fit_transform(y)
```

```python
ax = sns.regplot(x=x, y=y)
ax.set_xlabel(x_factor )
ax.set_ylabel(y_factor )
ax.set_title(evaluation)
plt.show()
```

```python
est = sm.OLS(y, x)
est2 = est.fit()
print(est2.summary())
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
r_sq = model.score(x, y)
print('R^2', r_sq)

print('intercept:', model.intercept_)

print('slope:', model.coef_)
```
